## ‚ùî **1. Analysis Question**

This project covers an exploratory data analysis (EDA), data preparation, modeling and comparison of algorithms, evaluation of performance metrics, and presentation of results. The focus is to identify the factors that influence the **propensity to pay the renewal premium in insurance policies** and prepare the dataset for a machine learning system that predicts this propensity. The guiding question is: ***What factors influence the likelihood of renewing insurance premiums, and how can the dataset be prepared, modeled, and evaluated to predict this probability and optimize incentives?***

### **1.1 Introduction**

The dataset used corresponds to data from an insurance company, focused on past transactions of policyholders and their propensity to renew premiums. This dataset is provided as **`insurance_company.csv`** in the case study.

It contains demographic information (age in days, monthly income, type of residence area), transactional information (percentage of premium paid in cash or credit, count of late premiums in 3-6/6-12/more than 12 months, underwriting score upon application, total premiums paid so far, sourcing channel) and policy details (monthly premium amount, renewal as binary target variable: 0 - Not Renewed, 1 - Renewed).

The objective of the exploratory analysis is to understand the fundamental characteristics of the dataset, identify patterns in premium renewals, detect anomalies (missing values, outliers) and prepare the dataset for modeling through cleaning, transformation, feature engineering, and variable selection. This will allow answering key questions, such as:

- What demographic, transactional, or policy factors influence renewal?
- What are the most common patterns of late payments and sourcing channels associated with non-renewals?
- How do income, age, and underwriting score impact the propensity to renew?

## üìù **6. EDA Conclusions**

### **Univariate Analysis**

#### **Findings and Conclusions for Numerical Variables**

1. **`perc_premium_paid_by_cash_credit`**

   - **Findings:** Range 0‚Äì1, mean ‚âà0.31, median 0.17, with 75% of customers below ‚âà0.54. Distribution strongly right-skewed: many values close to 0 and a long tail toward high values.

   - **Conclusion:** Most customers pay a moderate or low fraction of the premium by cash/credit, while a smaller group pays nearly 100%. This pattern suggests heterogeneous payment behaviors that may be associated with different levels of commitment or risk of non-renewal.

2. **`age_in_days`**

   - **Findings:** Range between ‚âà7,670 and ‚âà37,602 days (approx. 21 to 103 years), mean ‚âà18,847 days (‚âà51‚Äì52 years) and median ‚âà18,625 days. Distribution relatively wide but without suspicious extreme values.

   - **Conclusion:** The portfolio mainly covers middle-aged and older adults, consistent with life/long-term insurance policies. Age is a relevant variable for understanding renewal patterns and does not present data quality issues.

3. **`income`**

   - **Findings:** Very wide range: minimum ‚âà24,030 and maximum above 90 million. Very high standard deviation, indicating strong income outliers. The position measures (25%, 50%, 75%) are concentrated well below the maximum.

   - **Conclusion:** The income distribution is heavily skewed by a few customers with extremely high incomes. Robust treatment is required to prevent these outliers from distorting modeling.

4. **`count_3-6_months_late`**

   - **Findings:** Mean ‚âà0.25, median 0, maximum 13. Most customers (25%, 50% and 75% percentiles) have 0 late payments in this range; only a minority have multiple late payments.

   - **Conclusion:** 3‚Äì6 months late payments are infrequent, but when present they can be high, representing a higher-risk customer segment.

5. **`count_6-12_months_late`**

   - **Findings:** Mean ‚âà0.08, median 0, maximum 17. Almost all customers have no late payments in this range; those who do form a long tail.

   - **Conclusion:** 6‚Äì12 months late payments are even rarer, but their presence indicates prolonged non-payment behaviors that may be strongly linked to non-renewal.

6. **`count_more_than_12_months_late`**

   - **Findings:** Mean ‚âà0.06, median 0, maximum 11. Distribution highly concentrated at 0 late payments with some extreme cases.

   - **Conclusion:** More-than-12-months late payments are rare but critical. These customers represent a high-risk profile that can be key to identifying likely non-renewals.

7. **`application_underwriting_score`**

   - **Findings:** Mean ‚âà99.07, range between ‚âà91.9 and 99.89, with percentiles tightly concentrated around 99. Variability is low (std ‚âà0.74).

   - **Conclusion:** Most customers have a high, concentrated underwriting score, suggesting the company mainly works with a relatively good-risk profile. Despite low dispersion, small differences in this score are statistically significant against renewal, so it remains relevant for the model.

8. **`no_of_premiums_paid`**

   - **Findings:** Mean ‚âà10.86, median 10, range from 2 to 60 premiums paid. 50% of customers have paid between 7 and 14 premiums.

   - **Conclusion:** Most policyholders have a moderate history of regular payments, reflecting some tenure and commitment to the policy. This history can influence the decision to renew.

9. **`premium`**

   - **Findings:** Range between 1,200 and 60,000, mean ‚âà10,925 and median 7,500. Right-skewed distribution: many low/medium premiums and a small group with very high premiums.

   - **Conclusion:** The portfolio is mainly composed of moderate-amount policies, while high-premium policies constitute a small but revenue-relevant segment. This variable is critical for analyzing the financial impact of renewal and incentive strategies.

#### **Findings and Conclusions for Categorical Variables**

1. **`sourcing_channel`**

   - **Findings:** 5 channels (`A`, `B`, `C`, `D`, `E`). Channel `A` concentrates most customers (‚âà43,000), followed by `B` and `C`; `D` and `E` are minority channels, especially `E`.

   - Conclusion: Channel `A` is the main customer acquisition driver, so any incentive or commercial strategy should focus on it. Smaller channels like `E` may have niche roles or lower commercial efficiency.

### **7.2 Cleaning Outlier Data**

```python
numerical_cols = [col for col in data_clean.select_dtypes(include=[np.number]).columns if col != 'renewal']

# outlier summary for numerical variables
outlier_summary = {}

for col in numerical_cols:
    Q1, Q3 = data_clean[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR

    outliers = data_clean[(data_clean[col] < lower) | (data_clean[col] > upper)]

    outlier_summary[col] = {
        'Outliers': len(outliers),
        '(%)': round(len(outliers) / len(data_clean) * 100, 2),
        'Lower Limit': round(lower, 2),
        'Upper Limit': round(upper, 2)
    }

outlier_summary_df = pd.DataFrame(outlier_summary).T
print("\nOutlier summary for numerical variables:")
print(outlier_summary_df)
```

```python
# detailed view of the first 5 outliers per variable
for col in numerical_cols:
    if col in data_clean.columns:
        Q1, Q3 = data_clean[col].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
        outliers = data_clean[(data_clean[col] < lower) | (data_clean[col] > upper)]
        if not outliers.empty:
            print(f"\nOutliers in '{col}' (first 5):")
            print(outliers[[col]].head())
```

#### **Treatment of Numerical Outliers**

The results show a high proportion of outliers in `count_3-6_months_late` (16.22%), `count_6-12_months_late` (4.92%), and `count_more_than_12_months_late` (4.66%), which were retained without modification to preserve variance and avoid computational issues (means close to zero, since the IQR limits are 0 and most values are low or zero). The variables `income` (4.29%) and `premium` (5.66%) were treated using logarithmic transformation to reduce skewness due to their heavily skewed distributions (high upper limits: 468210 and 26400, with examples of extreme values). `application_underwriting_score` (4.74%) and `no_of_premiums_paid` (1.79%) were treated with capping at the IQR limits to minimize noise. `age_in_years` (0.06%) was also capped due to its very small proportion of outliers. `perc_premium_paid_by_cash_credit` (0% outliers) required no treatment.

```python
# log-transformation for 'income' and 'premium'
for col in ['income', 'premium']:
    data_clean[col] = np.log(data_clean[col] + 1)  # +1 to avoid log(0)
```

```python
# capping for 'application_underwriting_score', 'no_of_premiums_paid', and 'age_in_years'
for col in ['application_underwriting_score', 'no_of_premiums_paid', 'age_in_years']:
    Q1, Q3 = data_clean[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    data_clean[col] = np.clip(data_clean[col], lower, upper)
```

#### **Treatment of Categorical Outliers**

Invalid or irrelevant categories in categorical variables are considered outliers, as they can introduce noise into predictive models. In this dataset, no invalid or irrelevant categories were identified in `sourcing_channel` (values: `A`, `B`, `C`, `D`, `E`) or in `residence_area_type` (values: `Urban`, `Rural`), as all are consistent and relevant for the renewal analysis. Therefore, no category removal is performed. Other consolidations could be applied during feature engineering if needed to reduce cardinality (although in this case, cardinality is low: 5 and 2 respectively).

## ü§ñ **8. Modeling / Algorithm Comparison**

In this section, three binary classification algorithms (*Random Forest Classifier*, *Logistic Regression*, and *KNN Classifier*) are compared to predict the probability of renewal (`renewal`). The problem is inherently binary (renew = 1; not renew = 0), so classification models are used instead of continuous regression. These algorithms generate calibrated probabilities between 0 and 1, suitable for prioritizing customers, assigning incentives, or segmenting risk.

The balanced dataset (`df_insurance`) is used, processed with feature engineering, MinMaxScaler normalization, and SMOTE to correct imbalance. An 80/20 train/test split is applied, and each model is evaluated using classification metrics (*Accuracy*, *Precision*, *Recall*, *F1-score*) and probability-based metrics (*Log Loss*, *Brier Score*, and *ROC-AUC*).

### **Importance of the metrics**

The goal is to predict a client's likelihood of paying the renewal premium in order to design an incentive plan that maximizes the company‚Äôs net revenue. For this purpose, it is essential to **correctly identify customers with a low probability of renewal**, as they are the natural candidates for incentives and retention actions. Therefore, the most relevant metrics are:

- ***Recall* (class 0 ‚Äì non-renewal):** avoiding false negatives is critical, since losing a high-value customer has greater impact than contacting someone who was already planning to renew.
- ***ROC-AUC*:** evaluates the model's overall ability to distinguish between renewers and non-renewers, independent of thresholds.
- ***F1-score*:** a secondary metric useful in the balanced post-SMOTE scenario where both precision and recall must be considered.

The most important metric for this case is the *Recall* of class 0 (customers who do not renew), because the cost of failing to intervene in time is higher than the cost of intervening with a customer who already intended to renew. Consequently, the selected model will be the one that maximizes class-0 *Recall* while maintaining a competitive *ROC-AUC* and reasonable penalties in *log-loss*, ensuring a balance between performance and operational utility.

### **Model optimization**

Each algorithm will undergo an optimization process (Grid Search or manual hyperparameter tuning) to compare:

- Baseline versions vs. optimized versions  
- Changes in accuracy, recall, and ROC-AUC  
- Adjustments in regularization, depth, number of neighbors, etc.

This will support selecting the optimal model according to technical performance and alignment with business objectives.

### **9.1 Metrics Comparison**

In this subsection, a detailed comparison of the models is presented using the following metrics:

- ***Accuracy*:** Proportion of total correct predictions `(TP + TN) / (TP + TN + FP + FN)`. It measures overall model performance, but it can be misleading in balanced datasets if asymmetric errors are ignored. In this case, after SMOTE, it is useful as a baseline but not a priority because the cost of FN (losing churners) is high.

* ***Precision* (class 1):** Proportion of true positives among predicted positives `TP / (TP + FP)`. It indicates how reliable renewal predictions are; a high value avoids unnecessary incentives, but it is not the main focus.

* ***Recall* (class 1):** Proportion of detected true positives `TP / (TP + FN)`. It measures the model‚Äôs ability to capture actual renewals; secondary here, since the focus is on class 0.

* ***Recall* (class 0):** Proportion of true negatives detected for class 0 `TN / (TN + FP for class 0)`. Primary metric: captures actual churners for intervention; a high value minimizes FN (missed churn), directly impacting retention-driven revenue.

* ***F1-score* (class 1):** Harmonic mean of *Precision* and *Recall* for class 1 `(2 * Precision * Recall) / (Precision + Recall)`. It balances precision and completeness; useful post-SMOTE for balanced datasets, but secondary given the churn focus.

* ***Log-Loss*:** Logarithmic penalty for incorrect probabilities `-1/N * Œ£ [y log(p) + (1-y) log(1-p)]`. Evaluates probability calibration; a low value indicates reliable probabilities for prioritizing incentives (e.g., risk-based thresholds).

* ***Brier Score*:** Mean squared error of probabilities `Œ£ (p - y)^2 / N`. Similar to Log-Loss, it measures calibration; a low value ensures realistic probabilities, avoiding suboptimal decisions in incentive planning.

* ***ROC-AUC*:** Area under the ROC curve (TPR vs. FPR). Measures general class discrimination independently of thresholds; a high value indicates good ranking of customers by churn risk, essential for segmentation.
